## Checking Conditions for Regression

### (1) Linearity

```python
#All dependent variables must be linear 
plt.subplot(1,3,1)
plt.rcParams["figure.figsize"] = (15,3)
plt.scatter(data['Year'], data['Price'])
plt.title("Year and Price")
#Graph is not linear but exponential

plt.subplot(1,3,2)
plt.rcParams["figure.figsize"] = (15,3)
plt.scatter(data['EngineV'], data['Price'])
plt.title("EngineV and Price")
#Graph to spread in all directions

plt.subplot(1,3,3)
plt.rcParams["figure.figsize"] = (15,3)
plt.scatter(data['Mileage'], data['Price'])
plt.title("Mileage and Price")
#negative curve
```

```python
#We need to imporve this now. As discussed in notes, 
#applying logarithm is an easy fix for this

data['logPrice'] = np.log(data['Price'])
data.drop(inplace = True, axis =1, columns = ['Price'])
data.describe(include = 'all')
#Add a new column that takes log of all values in Price column
```

```python
plt.subplot(1,3,1)
plt.rcParams["figure.figsize"] = (15,3)
plt.scatter(data['Year'], data['logPrice'])
plt.title("Year and Price")

plt.subplot(1,3,2)
plt.rcParams["figure.figsize"] = (15,3)
plt.scatter(data['EngineV'], data['logPrice'])
plt.title("EngineV and Price")

plt.subplot(1,3,3)
plt.rcParams["figure.figsize"] = (15,3)
plt.scatter(data['Mileage'], data['logPrice'])
plt.title("Mileage and Price")
#Results are much better!
```

### (2) No Endogeneity

States that error and Price should NOT be co-related. In this case they are not, so this assumption is True

### (3) Normaility & Homoscedasticity

Errors should be Nomral, with 0 mean. For Homoscedasticity, we can see that there is not as the graphs are generally linearly distributed. Take Price and Year for example.

The amount of cars in early 90's is similar to that in the 2010's. Thus values are not being SPREAD abnormally. Since we applied Log, we already fixed that issue

### (4) No AutoCorrelation

This assumption is true as the data we are dealing with is not time series or panel data. Thus no serial correlation exists

### (5) Multicollinearity

There should be no co relation between any dependent variables. Note how 3rd Normal Form Normalized Databases in SQL also need this assumption to be True! That's why those data sets are easy to apply analysis tools on.

In our case, Year and Mileage can be co related a lot. After all, the older the car, the more likely it is to have been on the road more often! So let us check those two!

We will use statsmodels in this case

```python
columntocheck = data[['Mileage', 'EngineV', 'Year']]
newdf = pd.DataFrame()

newdf["VIF"] = [variance_inflation_factor(columntocheck.values, x) for x in range(columntocheck.shape[1])]
newdf.rename(index={0:'Mileage', 1:'EngineV', 2:'Year'}, inplace=True)
newdf
```

VIF = 1 means no multicollinearity

1<VIF<5 means perfectly ok

VIF>10 means unacceptable

Although there is set criteria, in some cases, a VIF of even 8 might be fine but in some case it must be less than 5! In our dataframe, we can see that year has a high VIF as we thought so.

Thus we will have to remove year!

```python
data.drop(axis = 1, columns = ['Year', 'Model'], inplace = True)
data.describe(include = 'all')
```

### Dealing with Categorical Data (Text)

In our example, brand, body, engine type etc are some examples. Normally we can create dummy's manually but here there are so many values that we need automation

For a feature with N categories, N-1 dummies are needed. This is because let's say we have 5 fruits: Apple, banana, mangoes, peach, and orange.

If we say the fruit is a banana, we rename it by 1. If we say the fruit is an orange, we rename it by 1 and so on. It is obvious that if all columns of the last 4 fruits are 0, then the fruit must be Apple, the first one on our list! If we create a separate dummy for it, it wil result in perfect multicollinearity which is not what we want!


```python
datanew = pd.get_dummies(data, drop_first = True)
#Notice how drop_first prevents the issue discussed above

datanew.head()
```

```python
datanew.columns.values
```

```python
newcol = [ 'logPrice', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol','Registration_yes']
```

```python
datanew = datanew[newcol]
datanew.head()
```

### Scaling Data

```python
#dividing data into dependent and independent variables
depend = datanew['logPrice']
indep = datanew.drop(axis =1, columns = ['logPrice'])
```

```python
depend.head()
```

```python
indep.head()
```

```python
scaler = StandardScaler()
scaler.fit(indep)
scaled_indep = scaler.transform(indep)
scaled_indep

#Note how we even scaled the dummy variables which is not a good practice though. A better method could have been to scale just numerical input data
```

### Train Test Split Model

```python
xtrain, xtest, ytrain, ytest = train_test_split(scaled_indep, depend, test_size = 0.2, random_state = 365)
#testsize = 0.2 separates 80% data for training and 20% for testing purposes
```

### Actual Regression

```python
reg = LinearRegression()
reg.fit(xtrain, ytrain)
#log linear regression created
```

```python
#We know how multiple linear regression equation is:
#y^ = b0 + b1x1 + b2x2 +.... so

yhat = reg.predict(xtrain)
plt.rcParams["figure.figsize"] = (6,3)
plt.scatter(yhat, ytrain)
plt.xlabel("Yhat - Predictions")
plt.ylabel("Ytrain - Target")
plt.xlim(6,13)
plt.ylim(6,13)
plt.plot(yhat, yhat, color = 'red', linewidth = 5)
plt.show()

#CAN YOU SEE HOW WELL THE REGRESSION FITTED!?
```

We know from one of our conditions for linear regression that errors must be normally distributed. Here, error = ytrain - yhat or how much different our predicted values are from actual values. This error should be normally distributed which as shown below, roughly is!

Since peak is more shifted towards the right, we can observe that our model tend to overestimate our targets and not underestimate them

```python
sns.displot(ytrain-yhat)
```

```python
#One of the last checks of calculating R2
#The 0.75 value says that our model predicts 75% of our data
reg.score(xtrain,ytrain)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
table = pd.DataFrame(indep.columns.values, columns = ['Features'])
table['Weights'] = reg.coef_
table['Intercept'] = reg.intercept_
table['R2'] = reg.score(xtrain, ytrain)
table
```

-ve weights mean that their impact cause car's price to decrease and vice versa. In the case of dummy variables though, remember we set audi as the benchmark, the column that we dropped? A positive weight for a dummy variable means that the respective category brand is more expensive or larger than the benchmark. A negative one indicates otherwise
